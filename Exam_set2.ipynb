```python
# Question 6 - show the graph of proportion of cars
cars = carsdf['brand'].unique()
num_car = []
for brand in cars:
    num = len(carsdf[carsdf['brand']==brand])
    num_car.append(num)
print(num_car)
fig = plt.figure(figsize =(10, 7))
plt.pie(num_car, labels = cars)
 
# show plot
plt.show()
```

```python
# xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx

proportions = carsdf.groupby('brand')['brand'].apply(lambda x: (x == 'car').mean())

# Create a line chart of the proportions by brand
proportions.plot(kind='line', figsize=(8, 6), grid=True, marker='o', color='b')

# Set the chart title and axis labels
plt.title('Proportions of Cars by Brand')
plt.xlabel('Brand')
plt.ylabel('Proportion')

# Display the chart
plt.show()
```
